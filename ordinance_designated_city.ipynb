```python
header_index = df_14.query('行番号 == "行番号"').index
df_14.drop(header_index, inplace=True)
```

```python
# 整数に変換できるように削除したので、整数変換
df_14['001:決算額'] = df_14['001:決算額'].astype(np.int64)
df_14['001:決算額'].dtype
```

```python
df_14.columns
```

```python
# 団体コードに謎の半角空白があるので、列名を変更する
df_14.rename(columns={'団体コード ':'団体コード'}, inplace=True)
```

```python
df_14.columns
```

```python
# 団体区分「1」が政令指定都市
df_14_政令市 = df_14.query('団体区分 == "1"')
df_14_政令市.head(3)
```

```python
# pivot_table: 各政令指定都市が1行になる
# index: 縦軸
# columns: 横軸
# values: 集計対象
# aggfunc: 集計方法, np.sum: 合計値
df_14_政令市_pivot = pd.pivot_table(df_14_政令市, 
                                 index=['団体コード', '団体名'], 
                                 columns='行名称',
                                 values='001:決算額', 
                                 aggfunc=np.sum)
df_14_政令市_pivot.head()
```

```python
list_columns = ['人件費', '物件費', '維持補修費', '扶助費', '補助費等', 
                '公債費', '積立金', '投資及び出資金・貸付金', '繰出金', 
                '前年度繰上充用金', '投資的経費', '歳出合計']
# 分析に必要な列の絞り込み
df_14_政令市_pivot = df_14_政令市_pivot[list_columns]
df_14_政令市_pivot.head()
```

```python
# 念の為、0行目の合計が歳出合計と一致しているか確認
print(df_14_政令市_pivot.iloc[0][list_columns][:11].sum() 
      == df_14_政令市_pivot.iloc[0][list_columns][11])
```

```python
# 表全体でも確認
print(df_14_政令市_pivot.iloc[:][list_columns].sum()[:11].sum() 
      == df_14_政令市_pivot.iloc[:][list_columns].sum()[11].sum())

print(df_14_政令市_pivot.iloc[:][list_columns].sum()[:11].sum())
print(df_14_政令市_pivot.iloc[:][list_columns].sum()[11].sum())
```

```python
# 性質別の決算額を歳出合計で割って、構成割合にする -> 歳出規模が異なる団体間の比較を可能に

# ピボットテーブルのコピー
df_14_政令市_pivot_ratio = df_14_政令市_pivot.copy()

# 各行の要素をその行の歳出合計で割る
for line in range(len(df_14_政令市_pivot_ratio)):
    df_14_政令市_pivot_ratio.iloc[line] = \
        (df_14_政令市_pivot_ratio.iloc[line] \
            / df_14_政令市_pivot_ratio.iloc[line]['歳出合計'] \
                * 100
                )
    
# indexを振り直す
df_14_政令市_pivot_ratio.reset_index(inplace=True)

df_14_政令市_pivot_ratio.head()
```

```python
# 0行目の割合の合計
iloc_0 = df_14_政令市_pivot_ratio.iloc[0][2:13].sum()
iloc_0
```

```python
# 分析の目的: どのような分野に予算を使用しているか分析する
```

```python
# 性質別に集計
x = np.array([])
for col in list_columns[:11]:
    tmp = df_14_政令市_pivot_ratio[col].values
    x = np.append(x, tmp)

x = x.reshape(11, 20)
x = tuple(x)
x   # array: 性質別, index: 政令市
```

```python
# 大阪市と新潟市のデータを集計
df_osaka = df_14_政令市_pivot_ratio.query('団体コード == "271004"')[list_columns].values
df_nigata = df_14_政令市_pivot_ratio.query('団体コード == "151009"')[list_columns].values

# 箱ひげ図の作成
plt.figure(figsize=(8, 6))
plt.boxplot(x, whis="range")
plt.scatter(range(1,12), df_osaka.reshape(-1)[:11], 
            color='r', label='大阪市')
plt.scatter(range(1,12), df_nigata.reshape(-1)[:11], 
            color='b', label='新潟市')
plt.xticks(range(1,12), list_columns[:11])
plt.title('政令指定都市 性質別構成比')
plt.legend(loc='upper right')
plt.show()
```

大阪
- 扶助費や公債費の割合が大きい
- 人件費、物件費、投資的経費の割合が小さい
- 投資的割合が小さい = 過去に行った投資の負担が大きく、将来への投資ができていない

新潟
- 扶助費の割合が小さい
- 人件費、物件費の割合が大きい
- 公的扶助の割合が小さい = 移民が医療や福祉に頼らず、就労による生活をしている
    - 医療機関や福祉施設が十分に整備されていない？
